```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import sys
from pprint import pprint
import requests
```

```python
data = pd.read_csv("no_nan_kunstveiling_nl_txmm.csv")
```

```python
data
```

```python
for indx, row in data.iterrows():
    response = requests.get(row["image-src"])
    print(str(indx).zfill(4))
    file_path = os.path.join("..", "data", f"{str(indx).zfill(4)}.png")
    data.at[indx, 'file_path'] = file_path
    with open(file_path, 'wb') as f:
        f.write(response.content)
```

```python
data.to_csv("no_nan_kunstveiling_nl_txmm.csv")
```
